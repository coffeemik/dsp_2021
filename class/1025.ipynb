```python
import thinkdsp as dsp
```

```python
cos_sig = dsp.CosSignal(freq=340, amp=1.0, offset=0)
sin_sig = dsp.SinSignal(freq=1700, amp=2.0, offset=0)
```

```python
cos_sig.plot()
dsp.decorate(xlabel='Time (s)')
```

```python
mix = sin_sig + cos_sig
mix.plot()
dsp.decorate(xlabel='Time (s)')
```

```python
waveA = sin_sig.make_wave(duration=3.0, start=0, framerate=11025)
waveA.make_audio()
```

```python
waveB = cos_sig.make_wave(duration=3.0, start=0, framerate=11025)
waveB.make_audio()
```

```python
waveC = mix.make_wave(duration=3.0, start=0, framerate=11025)
waveC.make_audio()
```
